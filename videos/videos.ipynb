Diferencia entre `valid`, `same` y `full` al hacer `correlate` o `convolve`


```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt

from matplotlib import animation
fig, ax = plt.subplots(2, 1, figsize=(6, 3), tight_layout=True, sharex=True)
x = np.arange(7)
y1 = np.array([1, 2, 3])
y2 = np.array([-1, -1])

def update(k): 
    ax[0].cla(); 
    ax[0].grid()
    ax[0].scatter(x[~(np.pad(y1, 2)==0)], y1)
    ax[0].set_xlim([-0.5, 6.5])
    y3 = np.roll(np.append(y2, [0, 0 ,0, 0, 0]), k)
    ax[0].scatter(x[~(y3==0)],y3[~(y3==0)])
    ax[1].cla();
    corr = np.correlate(y1, y2, mode='full')
    ax[1].scatter(x[1:-2], corr, marker='o', s=150, label='full')
    ax[1].scatter(x[1:-3], corr[:-1], marker='o',s=100, label='same')
    ax[1].scatter(x[2:-3], corr[1:-1], marker='o',s=50, label='valid')
    if k>0 and k<=len(corr):
        ax[1].scatter(k, corr[k-1], s=50, c='k', marker='x')
    ax[1].legend(loc=4)
    ax[1].grid()
    
anim = animation.FuncAnimation(fig, update, frames=6, interval=500, blit=True, repeat=False)
#anim.save('animation_same_valid_full.gif', dpi=80, writer='imagemagick')
```

```python
fig = plt.figure(figsize=(5, 5), tight_layout=True)
ax = fig.add_subplot(111, projection='3d') 
ax.set_proj_type('ortho')

x = np.linspace(-2*np.pi, 2*np.pi, 40)
X, Y = np.meshgrid(x, x); 
I = np.cos(X)*np.cos(Y);

def update(angle):
    im = ax.plot_surface(X, Y, I, shade=False, cmap=plt.cm.Greys); 
    true_angle = 45 + 45*angle/20
    if angle >=21:
        true_angle = 90
    ax.view_init(true_angle, true_angle); 
    ax.set_xlim3d([-2*np.pi, 2*np.pi]) 
    ax.set_ylim3d([-2*np.pi, 2*np.pi])
    return (im,)
    

anim=animation.FuncAnimation(fig, update, frames=31, interval=200, blit=True)
anim.save('clases/videos/image3dto2d.mp4')
```

```python
from scipy import fftpack

fig, ax = plt.subplots(1, 2, figsize=(8, 4), tight_layout=True); 

def update(k):
    x = np.arange(0, 80)
    X, Y = np.meshgrid(x, x)
    if k <= 39:
        img = np.cos(2.0*np.pi*X*k/80)
    elif k > 39 and k < 79:
        img = np.cos(2.0*np.pi*Y*(k-40)/80)
    else:
        img = np.cos(2.0*np.pi*X*(k-80)/80 + 2.0*np.pi*Y*(k-80)/80)
    # img = np.cos(2.0*np.pi*X*x_f/80) + np.cos(2.0*np.pi*Y*y_f/80)
    # img = np.cos(2.0*np.pi*X*x_f/80)*np.cos(2.0*np.pi*Y*y_f/80)
    S_img = np.abs(fftpack.fft2(img))    
    ax[0].cla()
    ax[1].cla()
    ax[1].set_title("Espectro de amplitud");  ax[0].set_title("Imagen")
    im = ax[1].imshow(fftpack.fftshift(S_img), cmap=plt.cm.Greys_r,
                      extent=[-40, 40, 40, -40])
    im = ax[0].imshow(img, cmap=plt.cm.Greys_r); 
    #return (im, )
    
anim=animation.FuncAnimation(fig, update, frames=120-1, interval=50, blit=True);
anim.save('clases/videos/dft2d-impulse.mp4')
```

```python
fig, ax = plt.subplots(1, 2, figsize=(8, 4), tight_layout=True); 

 
def update(pix_angle=0):
    img = np.zeros(shape=(80, 80));     
    for i in range(80):
        if pix_angle <= 80:
            img[i, int(i - 2*pix_angle*i/80 + pix_angle)] = 1 
        else:
            img[i, int(i - 2*(160-pix_angle)*i/80 +160 - pix_angle)] = 1 
        
    ax[0].cla()
    ax[1].cla()
    ax[1].set_title("Espectro de magnitud");  
    ax[0].set_title("Imagen")
    S_img = np.abs(fftpack.fft2(img))
    im = ax[1].imshow(fftpack.fftshift(S_img), cmap=plt.cm.Greys_r, 
                      extent=[-40, 40, 40, -40])
    im = ax[0].imshow(img, cmap=plt.cm.Greys_r); 
    #return (im, )

anim = animation.FuncAnimation(fig, update, frames=159, interval=50, blit=True);
anim.save('clases/videos/dft2d-line2.mp4')
#interact(update, pix_angle=IntSlider(min=0, max=159, value=0));
```

```python
fig, ax = plt.subplots(1, 2, figsize=(6, 3.5), tight_layout=True); 

def update(pix_angle=0):
    img = np.zeros(shape=(60, 60));
    for i in range(120):
        for k1 in range(-4, 4):
            for k2 in range(-4, 4):
                if np.abs(i+20*k2) < 60 and np.abs(20*k1 + int(i - 2*pix_angle*i/20 + pix_angle)) < 60:
                    img[i +20*k2, 20*k1 + int(i - 2*pix_angle*i/20 + pix_angle)] = 1 
    
    ax[0].matshow(img, cmap=plt.cm.Greys_r); 
    im= ax[1].matshow(img[20:-20,20:-20], cmap=plt.cm.Greys_r); 
    return (im, )

anim=animation.FuncAnimation(fig, update, frames=20, interval=500, blit=True)
anim.save('clases/videos/dft2d-line3.mp4')
#interact(update, pix_angle=IntSlider(min=0, max=20, value=0));
```

```python
fig, ax = plt.subplots(1, 2, figsize=(8, 4), tight_layout=True); 

def update(width=1):
    img = np.zeros(shape=(80, 80)); #S_img[0, 0] = 0.0;
    img[40-width:40+width, 40-width:40+width] = 1.0; 
    ax[1].set_title("Espectro de amplitud")
    S_img = np.abs(fftpack.fft2(img))
    im = ax[1].imshow(fftpack.fftshift(S_img), cmap=plt.cm.Spectral_r,
                      extent=[-40, 40, 40, -40])
    ax[0].set_title("Imagen")
    im = ax[0].imshow(img, cmap=plt.cm.Greys_r); 
    return (im, )

anim=animation.FuncAnimation(fig, update, frames=39, interval=100, blit=True)
anim.save('clases/videos/dft2d-rect.mp4')

#interact(update, width=IntSlider(min=1, max=39, value=0, description="TamaÃ±o"));
```

```python
#from ipywidgets import FloatSlider
fig, ax = plt.subplots(1, 2, figsize=(8, 4), tight_layout=True); 

def update(bandwidth=1.0):
    x = np.linspace(-100, 100, num=500)
    X, Y = np.meshgrid(x, x)
    img = np.exp(-0.5*(X**2 + Y**2)/(1+bandwidth)**2)
    ax[1].set_title("Espectro de amplitud")
    S_img = np.abs(fftpack.fft2(img))
    im = ax[1].imshow(fftpack.fftshift(S_img),  cmap=plt.cm.Spectral_r,
                      extent=[-250, 250, 250, -250])
    ax[0].set_title("Imagen")
    im = ax[0].imshow(img, cmap=plt.cm.Greys_r); 
    return (im, )
    
anim=animation.FuncAnimation(fig, update, frames=39, interval=100, blit=True)
anim.save('clases/videos/dft2d-gaussian.mp4')
#interact(update, bandwidth=FloatSlider(min=0.1, max=50.0, value=0, description="Ancho"));
```
